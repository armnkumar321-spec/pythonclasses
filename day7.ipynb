```python
import pandas as pd
df=pd.read_csv('student_data.csv')
df.groupby('study_hours')['exam_score'].mean()
```

```python
def get_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    else:
        return "F"
df['grade']=df['exam_score'].apply(get_grade)
```
