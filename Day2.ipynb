```python
lst=[1,2,3,'hello','c',[2,4]]
print(lst)
#append()
lst.append([2,3,4])
print(lst)
#extend
lst.extend([2,3,4])
print(lst)
#insert(index,value)
lst.insert(2,['a','b'])
print(lst)
#pop()
lst.pop()
print(lst)
#remove()
lst.remove(2)
print(lst)
#index()
print(lst.index(2))

#cound()
print(lst.count(3))



```

#lst=[];
#append()-> used to add list in last of the block;
#extend()->used to add single value in last list
#insert(index,value)->used to add vale or list between the list
#pop()->used to remove element from last
#remove(value)->first occurence only remove that value
#index(start,end)-> print value of index
#count()->count the given value

```python
#sum
lst=[2,3,4,9,5,6]
a=sum(lst)
print(a)
#sorted
b=sorted(lst)
print(b)
#max
c=max(lst)
print(c)
#min
d=min(lst)
print(d)
```
