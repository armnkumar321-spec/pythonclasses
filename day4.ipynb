### Column-wise Concatenation Example

To concatenate arrays column-wise, you use `np.concatenate` and set the `axis` parameter to `1`. This means the arrays are joined side-by-side along the second dimension.

```python


a = np.array([[10, 20],
                    [30, 40]])

b = np.array([[50, 60],
                    [70, 80]])

print(a)
print(b)


c = np.concatenate((a,b), axis=1)

print(c)
d = np.concatenate((a,b), axis=0)
print(d)

```

```python
import numpy as np
n=int(input("enter size"))
lst=[]
DAYS=['SUN','MON','TUE','WED','THU','FRI','SAT']
for i in range(n):
  score=int(input(f"enter score FOR THE {DAYS[i]}:"))
  lst.append(score)
# print(" scores:", lst)


player_scores = np.array(lst)
print("Player scores ", player_scores)

std_score=np.std(player_scores)
print("SD of the all scores :", std_score)

max_score=np.max(player_scores)
max_score_index = np.argmax(player_scores)
max_score_day = DAYS[max_score_index]
print(f"Max score is on {max_score_day}: {max_score}")

min_score=np.min(player_scores)
min_score_index=np.argmin(player_scores)
min_score_day=DAYS[min_score_index]
print(f"Min score is on {min_score_day}: {min_score}")

mean_score=np.mean(player_scores)
print("mean of scores : ",mean_score)

median_score=np.median(player_scores)

if median_score in player_scores:
    median_score_index = np.where(player_scores == median_score)[0][0]
    median_score_day = DAYS[median_score_index]
    print(f"Median score is on {median_score_day}: {median_score}")
else:
    print(f"Median score: {median_score} (This value is not present )3")
```
