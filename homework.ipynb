```python
import numpy as np
n=int(input("Enter number of items: "))
lst=[]
ITEMS=['Item 1','Item 2','Item 3','Item 4','Item 5','Item 6','Item 7']
for i in range(n):

  price=int(input(f"Enter price for {ITEMS[i]}:\n"))
  lst.append(price)
# print(" prices:", lst)


item_prices = np.array(lst)
print("Item prices ", item_prices)

std_price=np.std(item_prices)
print("SD of the all prices :", std_price)

max_price=np.max(item_prices)
max_price_index = np.argmax(item_prices)
max_price_item = ITEMS[max_price_index]
print(f"Max price is for {max_price_item}: {max_price}")

min_price=np.min(item_prices)
min_price_index=np.argmin(item_prices)
min_price_item = ITEMS[min_price_index]
print(f"Min price is for {min_price_item}: {min_price}")

mean_price=np.mean(item_prices)
print("Mean of prices : ",mean_price)

median_price=np.median(item_prices)

if median_price in item_prices:
    median_price_index = np.where(item_prices == median_price)[0][0]
    median_price_item = ITEMS[median_price_index]
    print(f"Median price is for {median_price_item}: {median_price}")
else:
    print(f"Median price: {median_price} (This value is not present )")
```
